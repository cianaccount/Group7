```python
from selenium import webdriver
import time
import pandas as pd
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--disable-gpu')
browser = webdriver.Chrome(chrome_options=chrome_options)
```

```python
song_urls = []
browser.get('https://music.163.com/#/discover/toplist?id=3778678')
browser.switch_to.frame('g_iframe')
items = browser.find_elements_by_xpath('//table[@class = "m-table m-table-rank"]/tbody/tr//span[@class = "txt"]/a')
for item in items:
    href = item.get_attribute('href')
    song_urls.append(href)
```

```python
def func_comments(i):
    comments_list=[]
    browser.get(song_urls[i])
    time.sleep(1)
    try:
        browser.switch_to.frame('g_iframe')
    except:
        pass
    song_name = browser.find_element_by_xpath('//div[@class = "tit"]/em')
    items = browser.find_elements_by_xpath('//div[@class = "cmmts j-flag"]//div[@class = "cntwrap"]/div/div[@class = "cnt f-brk"]')
    for item in items:
        comments_list.append(item.text.split('ï¼š')[1] + '\n')
    temp=pd.DataFrame(comments_list[0:15],columns=[song_name.text])
    return temp
```

```python
comments=pd.DataFrame()
for i in range(len(song_urls)):
    comments=pd.concat([comments,func_comments(i)],axis=1)
    percent=i/len(song_urls)*100
    print(percent,'% finished')
```

```python
comments.to_csv('comments.csv',encoding='utf_8_sig')
```
